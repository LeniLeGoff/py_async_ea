```python
import seaborn as sns
import pandas as pd

def load_fitness(foldername):
    fit_file = open(foldername + "/fitnesses")
    index_file = open(foldername + "/indexes")
    fit_lines = fit_file.readlines()
    index_lines = index_file.readlines()
    data = []
    iteration = 0
    for fit_l, index_l in zip(fit_lines,index_lines):
        for fit, index in zip(fit_l.split(","),index_l.split(",")):
            data.append([iteration,int(index),float(fit)])
        iteration+=1
    fit_file.close()
    index_file.close()
    return data
```

```python
fit_data_lines = load_fitness("/home/leni/git/py_async_ea/logs/awg/m2w_2023_9_5_15_52_48_618139_2274710/")
fit_data = pd.DataFrame(data=fit_data_lines,columns=["iteration","index","fitness"])
print(fit_data)
```

```python
sns.lineplot(data=fit_data,x="iteration",y="fitness")
```

```python
sns.lineplot(data=fit_data,x="iteration",y="fitness",estimator="std")
```
